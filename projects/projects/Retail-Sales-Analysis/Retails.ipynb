```python
#Importing of libraries 
import pandas as pd
import numpy as np
```

```python
# Reading of files
sales =pd.read_csv(r"C:\Users\juwer\Downloads\Ratails Analytics\sales_data.csv")
customers=pd.read_csv(r"C:\Users\juwer\Downloads\Ratails Analytics\customers.csv")
products=pd.read_csv(r"C:\Users\juwer\Downloads\Ratails Analytics\products.csv")
returns=pd.read_csv(r"C:\Users\juwer\Downloads\Ratails Analytics\returns.csv")
stores=pd.read_csv(r"C:\Users\juwer\Downloads\Ratails Analytics\stores.csv")
```

# Sales Data Cleaning

```python
sales.info()
```

```python
# Finding null values
sales.isnull().sum()
```

```python
# Filling of null values
sales['store_id'] = sales['store_id'].fillna(sales['store_id'].mode()[0])
```

```python
sales.isnull().sum()
```

```python
# Standardizing Formats
sales.columns = sales.columns.str.strip().str.title()
```

```python
# Finding duplicates
duplicates = sales[sales.duplicated(subset='Order_Id')]
duplicates.shape[0]
```

```python
# removing duplicates 
sales.drop_duplicates(subset=['Order_Id'], inplace=True)
```

```python
duplicates = sales[sales.duplicated(subset='Order_Id')]
duplicates.shape[0]
```

```python
# Sorting 
sales = sales.sort_values(by='Order_Id')
```

```python
# Fixing of Outliers
# Columns to check
cols = ['Quantity', 'Unit_Price', 'Discount_Pct', 'Total_Amount']

# Loop through each column
for col in cols:
    Q1 = sales[col].quantile(0.25)
    Q3 = sales[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Count outliers
    outlier_count = ((sales[col] < lower) | (sales[col] > upper)).sum()
    print(f"Outliers in '{col}': {outlier_count}")

    # Fix outliers by capping
    sales[col] = sales[col].clip(lower, upper)
```

```python
# Checking of data types
sales.dtypes
```

```python
# Changing of Order_Date datatype
sales['Order_Date'] = pd.to_datetime(sales['Order_Date'], errors='coerce')
```

```python
sales.dtypes
```

```python
# Caluclating profit
sales = sales.merge(products[['Product_Id', 'Cost_Price']], on='Product_Id', how='left')
sales['Profit'] = (sales['Total_Amount'] - (sales['Cost_Price'] * sales['Quantity'])).round(2)
```

```python
sales.describe()
```

```python
sales.info()
```

```python
sales.shape
```

```python
# Saving of cleaned data
sales.to_csv('sales_cleaned.csv',index=False)
```

# Customers Data Cleaning

```python
customers.head()
```

```python
customers.info()
```

```python
# Finding null values
customers.isnull().sum()
```

```python
# Filling null values
customers['age'] = customers['age'].fillna(customers['age'].median())
```

```python
customers.isnull().sum()
```

```python
# Standardizing Formats
customers.columns = customers.columns.str.strip().str.title()
```

```python
# Finding duplicates
duplicates = customers[customers.duplicated(subset='Customer_Id')]
duplicates.shape[0]
```

```python
# Removing duplicate values
customers = customers.drop_duplicates(subset='Customer_Id')
```

```python
duplicates = customers[customers.duplicated(subset='Customer_Id')]
duplicates.shape[0]
```

```python
# Sorting 
customers = customers.sort_values(by='Customer_Id')
```

```python
# Finding of outliers

Q1 = customers['Age'].quantile(0.25)
Q3 = customers['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = customers[(customers['Age'] < lower_bound) | (customers['Age'] > upper_bound)]
print(f"Number of outliers: {outliers.shape[0]}")
```

```python
# Finding datatypes
customers.dtypes
```

```python
# fixing of data type
customers['Signup_Date'] = pd.to_datetime(customers['Signup_Date'])
```

```python
# Checking of data types
customers.dtypes
```

```python
# Creating derived columns

def Age_group(age):
    if age < 25:
        return 'Youth'
    elif age < 40:
        return 'Adult'
    elif age < 60:
        return 'Mid-Age'
    else:
        return 'Senior'

customers['Age_group'] = customers['Age'].apply(Age_group)
```

```python
customers.describe()
```

```python
customers.shape
```

```python
# Saving of cleaned data
customers.to_csv('customers_cleaned.csv',index=False)
```

# Products Data Cleaning

```python
products.head()
```

```python
# Finding null values
products.isnull().sum()
```

```python
# Filling missing brand values with mode
products['brand']=products['brand'].fillna(products['brand'].mode()[0])
```

```python
products.isnull().sum()
```

```python
# Standardizing Formats
products.columns = products.columns.str.strip().str.title()
```

```python
# Finding duplicates
duplicates = products[products.duplicated(subset='Product_Id')]
duplicates.shape[0]
```

```python
# Dropping duplicates based on product_id, keeping the first occurrence
products = products.drop_duplicates(subset='Product_Id', keep='first')
```

```python
duplicates = products[products.duplicated(subset='Product_Id')]
duplicates.shape[0]
```

```python
# Sorting 
products = products.sort_values(by='Product_Id')
```

```python
# Defining numeric columns
cols = ['Cost_Price', 'Unit_Price', 'Margin_Pct']

# Loop through each column and count outliers
for col in cols:
    Q1 = products[col].quantile(0.25)
    Q3 = products[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_count = ((products[col] < lower) | (products[col] > upper)).sum()
    print(f"Outliers in '{col}': {outlier_count}")
```
